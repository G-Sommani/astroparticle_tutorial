## Point 1
Which Fermi sources are located within 2 degrees from the center of the region? Hint: You can look up the sources in the Fermi-LAT 4FGL catalog, that is in the repository under data/gll_psc_v23.fit.

```python
#Package to open fits files
import astropy.io.fits as pyfits
```

```python
#Open the fit file (check the position of this notebook in your system!)
fermi_list = pyfits.open('../data/gll_psc_v23.fit')
```

```python
#Explore the data in the file
fermi_list[1].columns
```

```python
#The central position
RA = 77.3625
DEC = 5.6989
```

```python
radius = 2.
```

```python
#Package to handle easier and faster the data
from astropy.table import Table
```

```python
#convert the section of the fits file in a table
point_src = Table(fermi_list[1].data)
```

```python
point_src
```

```python
import numpy as np
```

```python
#To see how many sources are within 2 degrees one of the fastest ways consists of applying a mask to the table
#here: definition of the mask
mask_radius = np.sqrt((RA - point_src['RAJ2000'])**2 + (DEC - point_src['DEJ2000'])**2) <= radius
```

```python
#application of the mask to the table
the_sources = point_src[mask_radius]
```

```python
#5 sources are within 2 degrees
the_sources
```

## Point 4
Use aperture photometry to derive the light curves of the two brightest sources starting at $100$ MeV (i.e. counting all photons above 100 MeV). Aperture photometry means to count all photons within a circle of given radius around the source position. Use a $0.5^\circ$ and a $3^\circ$ aperture radius. Use a 30-day time binning.

```python
#First, apply the energy filter
photons = photon_table[photon_table['ENERGY'] > 100]
```

```python
#The two brightest sources
source1 = the_sources[1]
source2 = the_sources[3]
```

```python
#Define the separation for the photometry
sep = 0.5
```

```python
#Define the masks for the photometry
mask1 = np.sqrt((photons['RA'] - source1['RAJ2000'])**2 + (photons['DEC'] - source1['DEJ2000'])**2) <= sep
mask2 = np.sqrt((photons['RA'] - source2['RAJ2000'])**2 + (photons['DEC'] - source2['DEJ2000'])**2) <= sep
```

```python
#Apply the masks
photons1 = photons[mask1]
photons2 = photons[mask2]
```

```python
#Check if the photons were correclty selected
plt.hist2d(photons1['RA'], photons1['DEC'], bins = 10)
plt.hist2d(photons2['RA'], photons2['DEC'], bins = 10)
plt.xlim(70, 85)
plt.ylim(-2,13)
plt.show()
```

```python
#Days in the time interval
max_t = max(photons['TIME'])
min_t = min(photons['TIME'])
days = (max_t - min_t)/(60*60*24)
days
```

```python
#Define the number of bins necessary to have 30 days per bin
bins = round(days/30)
bins
```

```python
#Compute the lightcurves
plt.figure(figsize = (10,3))
plt.hist(photons1['TIME'], bins = bins, histtype = 'step')
plt.hist(photons2['TIME'], bins = bins, histtype = 'step')
plt.show()
```

```python
#Repeate with an aperture radius of 3 degrees

sep = 3.

mask1 = np.sqrt((photons['RA'] - source1['RAJ2000'])**2 + (photons['DEC'] - source1['DEJ2000'])**2) <= sep
mask2 = np.sqrt((photons['RA'] - source2['RAJ2000'])**2 + (photons['DEC'] - source2['DEJ2000'])**2) <= sep

photons1 = photons[mask1]
photons2 = photons[mask2]

plt.figure(figsize = (10,3))
plt.hist(photons1['TIME'], bins = bins, histtype = 'step')
plt.hist(photons2['TIME'], bins = bins, histtype = 'step')
plt.show()
```

```python
#With 3 degrees of aperture radius the tue setis of photons overlap
plt.hist2d(photons1['RA'], photons1['DEC'], bins = 10)
plt.hist2d(photons2['RA'], photons2['DEC'], bins = 10)
plt.xlim(70, 85)
plt.ylim(-2,13)
plt.show()
```
